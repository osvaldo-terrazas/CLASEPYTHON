```python
##Importamos las librerias: numpy y matplotlib
#Esta es una prueba de modificación de código
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definimos 2 arreglos de dimensión 100
ys = 200 + np.random.randn(100)
x = np.arange (0, 100, 1)
```

```python
#Graficamos los arreglos en un plano 2D
plt.plot(x, ys, '-')
plt.fill_between(x, ys,195, facecolor='r', alpha=0.6)

#Agregamos título a la gráfica
plt.title("Sample Visualization")
plt.xlabel("eje de las x", size = 16,)
plt.ylabel("eje de las y", size = 16,)
plt.show()
```
